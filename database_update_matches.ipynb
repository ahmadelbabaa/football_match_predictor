```python
import pandas as pd
import numpy as np
```

```python
# insert here the file to upload
matches_file = 'match_results_big5_2021.csv'
```

### importing data and updating matches table

```python
new_matches = pd.read_csv(f'data/{matches_file}')
new_matches.head()
```

```python
new_matches.drop(['Unnamed: 0','Gender','Country', 'Round', 'Venue','Referee', 'Notes','Attendance', 'MatchURL'], axis=1, inplace=True)
# matches['match_id'] = np.arange(len(matches))
new_matches.head()
```

```python
seasons = pd.read_csv('tables/seasons.csv')
seasons
```

```python
leagues = pd.read_csv('tables/leagues.csv')
leagues
```

```python
new_matches = pd.merge(new_matches, seasons,on=['Season_End_Year'], how='left')
new_matches = pd.merge(new_matches, leagues,on=['Competition_Name'], how='left')
new_matches.drop(['Season_End_Year', 'Competition_Name'], axis=1, inplace=True)
new_matches.head()
```

```python
teams=pd.read_csv('tables/teams.csv')
teams.head()
```

```python
new_matches = pd.merge(new_matches, teams[['Squad','team_id']] ,left_on='Home', right_on='Squad', how='left')
new_matches.rename(columns={'team_id':'home_id'}, inplace=True)
new_matches= pd.merge(new_matches, teams[['Squad','team_id']] ,left_on='Away', right_on='Squad', how='left')
new_matches.rename(columns={'team_id':'away_id'}, inplace=True)
new_matches.drop(['Home', 'Away', 'Squad_x', 'Squad_y'],axis=1, inplace=True)
new_matches.head()
```

```python
# set match result
result = np.zeros(len(new_matches), dtype = str)

for i in range(len(new_matches)):
    if np.array(new_matches['HomeGoals'])[i] == np.array(new_matches['AwayGoals'])[i]:
        result[i] = 'x'
    elif np.array(new_matches['HomeGoals'])[i] < np.array(new_matches['AwayGoals'])[i]:
        result[i] = 2
    else:
        result[i] = 1
new_matches['result'] = result
new_matches
```

```python
matches = pd.read_csv('tables/matches.csv')
matches
```

```python
ids = np.arange(len(matches), len(matches)+len(new_matches))
new_matches['match_id'] = ids
new_matches
```

```python
new_matches.loc[1:]
```

```python
num_row_uploaded=0

for i in range(len(new_matches)):
    if new_matches.loc[i]['match_id'] not in np.array(matches['match_id']):
        matches = pd.concat([matches, new_matches.loc[i:i]], axis=0, ignore_index=True)
        num_row_uploaded +=1
        
        
print(f'Uploaded {num_row_uploaded} rows!')
matches.to_csv('tables/matches.csv', index=False)
matches
```
