```python
import pandas as pd
import numpy as np
```

### importing data and organizing data in tables

#### Teams table

```python
teams = pd.read_csv('data/teams.csv')
teams
```

```python
leagues = pd.read_csv('data/leagues.csv')
leagues.to_csv('tables/leagues.csv', index=False)
```

```python
# saving the table
teams.to_csv('tables/teams.csv', index=False)
```

#### Season Table

```python
seasons = pd.read_csv('data/seasons.csv')
seasons.to_csv('tables/seasons.csv', index=False)
seasons
```

#### Team_Season table

```python
defense = pd.read_csv('data/laliga_2020_defense.csv')
team_season = defense[['Squad','Season_End_Year','Num_Players','Mins_Per_90']][defense['Team_or_Opponent']=='team']
team_season = pd.merge(team_season,seasons,on=['Season_End_Year'], how='left')
team_season = pd.merge(team_season, teams[['Squad', 'team_id']], on=['Squad'], how='left')
team_season.drop([ 'Season_End_Year', 'Squad'], axis=1, inplace=True)
team_season
```

```python
team_season.to_csv('tables/team_season.csv', index=False)
```

#### Defense table

```python
defense = pd.read_csv('data/laliga_2020_defense.csv')
defense.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
defense.loc[defense['Team_or_Opponent'] == 'opponent', 'Squad'] = defense.loc[defense['Team_or_Opponent'] == 'opponent', 'Squad'].str[3:]

defense = pd.merge(defense, teams[['Squad', 'team_id']], on=['Squad'], how='left')
defense = pd.merge(defense, seasons,on=['Season_End_Year'], how='left')
defense.drop(['Season_End_Year', 'Squad', 'Mins_Per_90', 'Num_Players', 'Gender', 'Country', 'Competition_Name'], axis=1, inplace = True)
defense.head()
```

```python
defense.to_csv('tables/defense.csv', index= False)
```
