```python
import pandas as pd
import numpy as np
```

### importing datasets and data cleaning

```python
players = pd.read_csv('data/soccer.players.csv')
players.drop('_id', axis=1, inplace=True)
```

```python
players.head()
```

```python
players['player_id'] = np.arange(len(players))
players.head()
```

```python
players_defense = pd.read_csv('data/player_defense2023.csv')
players_defense.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_defense = pd.merge(players_defense, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_defense.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_defense.head()
```

```python
players_gca = pd.read_csv('data/player_gca2023.csv')
players_gca.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_gca = pd.merge(players_gca, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_gca.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_gca.head()
```

```python
players_keeper_adv = pd.read_csv('data/player_keeper_adv2023.csv')
players_keeper_adv.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_keeper_adv = pd.merge(players_keeper_adv, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_keeper_adv.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_keeper_adv.describe()
```

```python
players_shooting = pd.read_csv('data/player_shooting2023.csv')
players_shooting.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_shooting = pd.merge(players_shooting, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_shooting.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_shooting.head()
```

```python
players_passing = pd.read_csv('data/player_passing2023.csv')
players_passing.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_passing = pd.merge(players_passing, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_passing.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_passing.head()
```

```python
players_passtype = pd.read_csv('data/player_passtype2023.csv')
players_passtype.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_passtype = pd.merge(players_passtype, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_passtype.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_passtype.head()
```

```python
players_possession = pd.read_csv('data/player_possession2023.csv')
players_possession.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_possession = pd.merge(players_possession, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_possession.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_possession.head()
```

```python
players_keepers = pd.read_csv('data/player_keepers2023.csv')
players_keepers.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
players_keepers = pd.merge(players_keepers, players[['Squad', 'Player', 'player_id']], on=['Squad', 'Player'], how='left')
players_keepers.drop(['Season_End_Year', 'Squad', 'Comp', 'Player', 'Nation', 'Pos', 'Age', 'Born','Url'], axis=1, inplace = True)
players_keepers.head()
players_keepers.describe()
```
