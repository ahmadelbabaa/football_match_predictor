```python
import pandas as pd
import numpy as np
```

```python
#files to update
#new season year
year = 2021 # only data to be modified

# files
defense_file = f'laliga_{year}_defense.csv'
gca_file = f'laliga_{year}_goal_shot_creation.csv'
keeper_adv_file = f'laliga_{year}_keeper_adv.csv'
shooting_file = f'laliga_{year}_shooting.csv'
passing_file = f'laliga_{year}_passing.csv'
passtype_file = f'laliga_{year}_pass_types.csv'
possession_file = f'laliga_{year}_possession.csv'
keepers_file = f'laliga_{year}_keepers.csv'
```

```python
# updating teh teams
defense = pd.read_csv(f'data/{defense_file}')
teams = pd.read_csv('tables/teams.csv')
leagues = pd.read_csv('tables/leagues.csv')

new_squads=np.array(defense[['Squad', 'Competition_Name']][defense['Team_or_Opponent']=='team'])
squads = np.array(teams['Squad'])

# adding to teams table the teams not in it
for team in new_squads:
    if team[0] not in squads:
        team_id = teams.iloc[-1][1]+1
        league = int(leagues['league_id'][leagues['Competition_Name']==team[1]])
        teams.loc[len(teams)] = [team[0], team_id, league]

# saving the table
teams.to_csv('tables/teams.csv', index=False)
```

#### Season Table

```python
seasons = pd.read_csv('tables/seasons.csv')
seasons
```

```python
# to  uodate the season table
if (year-2023) not in seasons['season_id']:
    new = pd.DataFrame({'season_id':[year-2023],'Season_End_Year':f'{year}'})
    seasons = pd.concat([seasons, new], axis=0, ignore_index=True)
    seasons.to_csv('tables/seasons.csv', index=False)
    print('Updated!')
    seasons
```

#### Team_Season table

```python
teams = pd.read_csv('tables/teams.csv')

defense = pd.read_csv(f'data/{defense_file}')
team_season = defense[['Squad','Season_End_Year','Num_Players','Mins_Per_90']][defense['Team_or_Opponent']=='team']
team_season = pd.merge(team_season,seasons,on=['Season_End_Year'], how='left')
team_season = pd.merge(team_season, teams[['Squad', 'team_id']], on=['Squad'], how='left')
team_season.drop([ 'Season_End_Year', 'Squad'], axis=1, inplace=True)
team_season.head()
```

```python
new_year = team_season.loc[0]['season_id']
old_team_season = pd.read_csv('tables/team_season.csv')
if new_year not in np.array(old_team_season['season_id']):
    team_season = pd.concat([old_team_season, team_season], axis=0, ignore_index=True)
    team_season.to_csv('tables/team_season.csv', index=False)
    print('Updated!')
team_season
```

#### Defense table

```python
defense = pd.read_csv(f'data/{defense_file}')
defense.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
defense.loc[defense['Team_or_Opponent'] == 'opponent', 'Squad'] = defense.loc[defense['Team_or_Opponent'] == 'opponent', 'Squad'].str[3:]
defense = pd.merge(defense, teams[['Squad', 'team_id']], on=['Squad'], how='left')
defense = pd.merge(defense, seasons,on=['Season_End_Year'], how='left')
defense.drop(['Season_End_Year', 'Competition_Name', 'Gender', 'Country', 'Mins_Per_90', 'Num_Players'], axis=1, inplace = True)
defense
```

```python
defense_year = defense.loc[0]['season_id']
old_defense = pd.read_csv('tables/defense.csv')
if defense_year not in np.array(old_defense['season_id']):
    defense = pd.concat([old_defense, defense], axis=0, ignore_index=True)
    defense.to_csv('tables/defense.csv', index=False)
    print('Updated!')
defense
```

### creating team stat table

```python
teams = pd.read_csv('tables/teams.csv')
team_season = pd.read_csv('tables/team_season.csv')
teams_stat = pd.merge(team_season,teams, on=['team_id'], how='left')
teams_stat
```

```python
teams_stat = pd.merge(teams_stat,seasons, on=['season_id'], how='left')
# teams_stat.drop('Season_End_Year', axis=1, inplace = True)
teams_stat
```

```python
defense = pd.read_csv('tables/defense.csv')
teams_stat = pd.merge(teams_stat, defense, on=['team_id', 'season_id'], how='left')
teams_stat.drop('id', axis=1, inplace=True)
teams_stat
```

```python
gca = pd.read_csv('tables/gca.csv')
teams_stat = pd.merge(teams_stat,gca, on=['team_id', 'Team_or_Opponent', 'season_id'], how='left')
teams_stat.drop('id', axis=1, inplace=True)
teams_stat.head()
```

```python
keepers = pd.read_csv('tables/keepers.csv')
teams_stat = pd.merge(teams_stat,keepers, on=['team_id', 'Team_or_Opponent', 'season_id'], how='left')
teams_stat.drop('id', axis=1, inplace=True)
teams_stat.head()
```

```python
keeper_adv = pd.read_csv('tables/keeper_adv.csv')
teams_stat = pd.merge(teams_stat,keeper_adv, on=['team_id', 'Team_or_Opponent', 'season_id'], how='left')
teams_stat.drop('id', axis=1, inplace=True)
teams_stat.head()
```

```python
shooting = pd.read_csv('tables/shooting.csv')
teams_stat = pd.merge(teams_stat,shooting, on=['team_id', 'Team_or_Opponent', 'season_id'], how='left')
teams_stat.drop('id', axis=1, inplace=True)
teams_stat.head()
```

```python
passing = pd.read_csv('tables/passing.csv')
teams_stat = pd.merge(teams_stat,passing, on=['team_id', 'Team_or_Opponent', 'season_id'], how='left')
teams_stat.drop('id', axis=1, inplace=True)
teams_stat.head()
```
