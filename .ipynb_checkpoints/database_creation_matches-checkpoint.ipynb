```python
import pandas as pd
import numpy as np
```

### importing data and creating matches table

```python
matches = pd.read_csv('data/match_results_big5_2020.csv')
matches.head()
```

```python
matches.drop(['Unnamed: 0','Gender','Country', 'Round', 'Venue','Referee', 'Notes','Attendance', 'MatchURL'], axis=1, inplace=True)
matches['match_id'] = np.arange(len(matches))
matches.head()
```

```python
seasons = pd.read_csv('tables/seasons.csv')
seasons
```

```python
leagues = pd.read_csv('tables/leagues.csv')
leagues
```

```python
matches = pd.merge(matches, seasons, on=['Season_End_Year'], how='left')
matches = pd.merge(matches, leagues, on=['Competition_Name'], how='left')
matches.drop(['Season_End_Year', 'Competition_Name'], axis=1, inplace=True)
matches.head()
```

```python
teams=pd.read_csv('tables/teams.csv')
teams.head()
```

```python
matches = pd.merge(matches, teams[['Squad','team_id']] ,left_on='Home', right_on='Squad', how='left')
matches.rename(columns={'team_id':'home_id'}, inplace=True)
matches= pd.merge(matches, teams[['Squad','team_id']] ,left_on='Away', right_on='Squad', how='left')
matches.rename(columns={'team_id':'away_id'}, inplace=True)
matches.drop(['Home', 'Away', 'Squad_x', 'Squad_y'],axis=1, inplace=True)
matches.head()
```

```python
# set match result
result = np.zeros(len(matches), dtype = str)

for i in range(len(matches)):
    if np.array(matches['HomeGoals'])[i] == np.array(matches['AwayGoals'])[i]:
        result[i] = 'x'
    elif np.array(matches['HomeGoals'])[i] < np.array(matches['AwayGoals'])[i]:
        result[i] = 2
    else:
        result[i] = 1
matches['result'] = result
matches
```

```python
matches.to_csv('tables/matches.csv', index=False)
```
